```python


from datetime import datetime, timedelta

# Task 1
from pprint import pprint

list_of_dates = [
    'Wednesday, October 2, 2002',
    'Friday, 11.10.13',
    'Thursday, 18 August 1977',
]


def dates_format_conversion(dates):
    for date in dates:
        try:
            print(datetime.strptime(date, "%A, %B %d, %Y").date())
        except ValueError:
            try:
                print(datetime.strptime(date, "%A, %m.%d.%y").date())
            except ValueError:
                try:
                    print(datetime.strptime(date, "%A, %d %B %Y").date())
                except ValueError:
                    print('Неизвестный формат...')
                    continue


dates_format_conversion(list_of_dates)

# Task 2
stream = [
    "2018-04-02",
    "2018-02-29",
    "2018-19-02",
]


def dates_checking(dates):
    new_stream = {}
    for date in dates:
        try:
            new_stream[date] = [datetime.strptime(date, "%Y-%m-%d").date(), True]
        except ValueError:
            new_stream[date] = [False]
    return new_stream


pprint(dates_checking(stream), sort_dicts=False)


# Task 3
def date_range(start_date, end_date):
    try:
        start_datetime = datetime.strptime(start_date, "%Y-%m-%d")
        end_datetime = datetime.strptime(end_date, "%Y-%m-%d")
        dates_difference = (end_datetime - start_datetime).days
        if dates_difference > 0:
            range_list_of_date = [start_date, end_date]
            new_list_date = start_datetime.date()
            cycle = 0
            while cycle < (dates_difference - 1):
                new_list_date += timedelta(days=1)
                range_list_of_date.insert(-1, new_list_date.strftime("%Y-%m-%d"))
                cycle += 1
        else:
            return []
        return range_list_of_date
    except ValueError:
        return []


print(date_range("2020-02-26", "2020-03-03"))
print(date_range("2020-26-02", "2020-03-03"))
print(date_range("2020-02-26", "2020-28-03"))

# Task 4
DEFAULT_USER_COUNT = 3


def delete_and_return_last_user(region, default_list=['A100', 'A101', 'A102']):
    try:
        default_list.remove(default_list[-1])
        return default_list[-1]
    except IndexError:
        print('Список закончелся, нечего возвращать...')


print(delete_and_return_last_user(1))
print(delete_and_return_last_user(1))
delete_and_return_last_user(1)

# Ошибка IndexError означала, что мы пытаемся обратиться к несущесвтующему
# индексу, когда возвращаем результат return default_list[DEFAULT_USER_COUNT-2],
# по сути здесь был задан всегда один и тот же индекс массива,
# а не последний, как требуется по заданию...
```
